```python
import pandas as pd
```

```python
us_states = pd.read_csv('data/us-states.csv')
us_states = us_states[['state', 'date', 'cases']]
us_states['date'] = pd.to_datetime(us_states['date'])

# Remove unneeded parts of the US
us_states = us_states[us_states['state'] != 'Northern Mariana Islands']
us_states = us_states[us_states['state'] != 'Guam']
us_states = us_states[us_states['state'] != 'Virgin Islands']
us_states = us_states[us_states['state'] != 'Puerto Rico']

us_states['date'] = us_states['date'].dt.month
us_states = us_states.groupby(['state', 'date']).max()
for state in us_states.index.get_level_values(0).unique():
    us_states.loc[(state, 0), :] = 0
    us_states.loc[(state, 11), :] = 0
us_states = us_states.reset_index()
us_states = us_states.sort_values(['state', 'date'])
us_states['monthly cases'] = us_states['cases'].diff()
us_states = us_states[(us_states['date'] > 0) & (us_states['date'] < 11)]
us_states = us_states.drop('cases', axis=1).set_index(['state', 'date'])

for state in us_states.index.get_level_values(0).unique():
    for month in range(1, 11):
        if not ((us_states.index.get_level_values(0) == state) & (us_states.index.get_level_values(1) == month)).any():
            us_states.loc[(state, month), :] = 0
us_states = us_states.reset_index().sort_values(['state', 'date']).reset_index(drop=True)
```

```python
dataframe = us_states.copy()
```

```python
us_states = pd.read_csv('data/us-states.csv')
us_states = us_states[['state', 'cases']]

# Remove unneeded parts of the US
us_states = us_states[us_states['state'] != 'Northern Mariana Islands']
us_states = us_states[us_states['state'] != 'Guam']
us_states = us_states[us_states['state'] != 'Virgin Islands']
us_states = us_states[us_states['state'] != 'Puerto Rico']
us_states = us_states.groupby('state').max().reset_index()
us_states['cases normalized'] = us_states['cases'] / us_states['cases'].sum()
us_states = us_states.set_index('state')
```

```python
dataframe['total cases'] = 0
dataframe['normalized total cases'] = 0
```

```python
us_states
```

```python
for state in dataframe['state'].unique():
    dataframe.loc[dataframe['state'] == state, 'total cases'] = us_states.loc[state]['cases']
    dataframe.loc[dataframe['state'] == state, 'normalized total cases'] = us_states.loc[state]['cases normalized']
```

```python
dataframe.to_csv('data/filtered-data.csv', index=False)
```
